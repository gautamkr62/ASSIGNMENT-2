```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

Experimental test 2100 participants 13 to 65
half under 65 and half equal or above
95 $ patients under 65 experince no side effect
95 % participitants above 65 experience side effects


```python
for i in range(50):
  # 5 % got side effect
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(1)
  # 5 % didn't got any side effect
  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(0);

for i in range(1000):
  # 95 % didn't got any side effect
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(0)
  # 95 % older got side effect
  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(1)

```

```python
for i in train_samples:
  print (i)
for i in train_labels:
  print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
# just rescaling age from 13 to 100 to 0 to 1 using sickit learn
scaler = MinMaxScaler(feature_range= (0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
  print(i)
```

```python
# but why scaling is usefule ? c
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
physical_devices = tf.config.experimental.list_physical_devices('GPU')
print("Num GPUs Available: ", len(physical_devices))
tf.config.experimental.set_memory_growth(physical_devices[0],True)
```

```python
# Sequential model are the simplest model can be thought of linear stack
model = Sequential([
    # there is also a input layer which is hidden layer.
    # inpput layer is not defined explicitly
    Dense(units = 16, input_shape=(1,),activation = 'relu'), # this first layer is actually the second layer, this is the first hidden layer
    Dense(units = 32, activation= 'relu'),
    Dense(units = 2, activation='softmax')
    # dense means fully connected neural network which is generally considered the best
    # Softmax function is generally used at the last layer to calculate probablity. It is widely used in last.
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer = Adam(learning_rate = 0.0001),
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

```

```python
model.fit(x = scaled_train_samples, y = train_labels, batch_size = 10, epochs = 30 , shuffle = True,verbose = 2 )

# what is meant by shuffle = True in model.fit ?

# verbose is the level of output.
```
