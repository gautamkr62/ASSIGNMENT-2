```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

Experimental test 2100 participants 13 to 65
half under 65 and half equal or above
95 $ patients under 65 experince no side effect
95 % participitants above 65 experience side effects


```python
for i in range(50):
  # 5 % got side effect
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(1)
  # 5 % didn't got any side effect
  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(0);

for i in range(1000):
  # 95 % didn't got any side effect
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(0)
  # 95 % older got side effect
  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(1)

```

```python
for i in train_samples:
  print (i)
for i in train_labels:
  print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
# just rescaling age from 13 to 100 to 0 to 1 using sickit learn
scaler = MinMaxScaler(feature_range= (0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
  print(i)
```

```python
# but why scaling is usefule ? c
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
physical_devices = tf.config.experimental.list_physical_devices('GPU')
print("Num GPUs Available: ", len(physical_devices))
# tf.config.experimental.set_memory_growth(physical_devices[0],True)

```

```python
# Sequential model are the simplest model can be thought of linear stack
model = Sequential([
    # there is also a input layer which is hidden layer.
    # inpput layer is not defined explicitly
    Dense(units = 16, input_shape=(1,),activation = 'relu'), # this first layer is actually the second layer, this is the first hidden layer
    Dense(units = 32, activation= 'relu'),
    Dense(units = 2, activation='softmax')
    # dense means fully connected neural network which is generally considered the best
    # Softmax function is generally used at the last layer to calculate probablity. It is widely used in last.
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer = Adam(learning_rate = 0.0001),
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

```

```python
model.fit(x = scaled_train_samples, y = train_labels, batch_size = 10, epochs = 30 , shuffle = True,verbose = 2 )

# what is meant by shuffle = True in model.fit ?

# verbose is the level of output.
```

```python
# use tensorflow to create the validation set on the fly during training.
 # but what is a validation set ?
# How well our model is generalizing by introducing a validation set.
# to create a validation set before training we can choose a subset of the training set labeled as valdation data..
# In each epoch after the model has gone through it will take what it learned from the training data and validate by predicting on the data in the validation set.
# using what it's learn from the training data there
# During running of the program. We will not only see the accuracy and loss of training set but also validation set.
# It's also important to understand that model is learning only through training data, it's not taking the validation set in account during training
# It's just say how our model is good at generalizing things.
# Knowing this information will allow us to check if there is overfitting or not in the model.
# overfiiting means learning well but unable to generalize the data it has seen before.
```

```python
# How to create a work with validation set using sequential model ?
# create a validation dataset from the training_set.
# then we will pass this dataset to the model when we trian it using the standard function
```

```python
model.fit(x = scaled_train_samples, y = train_labels, validation_split = 0.1, batch_size = 10, epochs = 30,shuffle = True, verbose = 2)
```

#Conufsion Matrix
Now we will visually see. How well our model is doing.

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true = test_labels, y_pred = rounded_predictions)
```

```python
def plot_confusion_matrix(cm,classes,
                          normalize = False,
                          title = 'Confusion matrix',
                          cmap = plt.cm.Greens):
  plt.imshow(cm, interpolation = 'nearest', cmap = cmap)
  plt.title(title)
  plt.colorbar()
  tick_marks = np.arange(len(classes))
  plt.xticks(tick_marks, classes, rotation = 45)
  plt.yticks(tick_marks, classes)

  if normalize:
    cm = cm.astype('float')/ cm.sum(axis= 1)[:, np.newaxis]
    print("Normalized confusion matrix")
  else:
    print("Confusion matrix, without normalization")

  print(cm)

  thresh = cm.max() / 2
  for i, j in itertools.product(range(cm.shape[0]) ,range(cm.shape[1])):
    plt.text(j,i,cm[i,j],
             horizontalalignment = "center",
             color = "red" if cm[i,j] > thresh else "black")
  plt.tight_layout()
  plt.ylabel('True_label')
  plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects', 'had_side_effects']
plot_confusion_matrix(cm = cm, classes = cm_plot_labels, title = 'Confusion Matrix')
```
